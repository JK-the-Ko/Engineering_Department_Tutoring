**Matplotlib**

Library to plot graph and visualize phenomenon (e.g. image)

```
import numpy as np #개발자들 사이의 약속 (단축어)
import matplotlib.pyplot as plt #개발자들 사이의 약속 (단축어)
```

* In order to plot a graph, the dimension of variables should be the same.

```
x = np.array([1,2,3]) #3
y = np.array([1,2,3,4]) #4
```

```
plt.plot(x, y) #Different Dimension
```

```
x = np.array([1,2,3,4]) #4
```

```
plt.plot(x, y) #Same Dimension
```

* In order to plot a graph, one needs to use **plt.plot(variable, variable)** method

* In order to show a graph, one needs to use **plt.show()** method

```
x = np.arange(0, 10, 0.001) #Independent Variable
y = np.sin(x) #Dependent Variable f(x)
```

```
print(type(y))
```

```
print(x.shape, y.shape)
```

```
plt.plot(x, y)
plt.show()
```

```
x = np.arange(0, 10, 1) #Independent Variable
y = np.sin(x) #Dependent Variable f(x)
```

```
plt.plot(x, y)
plt.show()
```

* Other features of matplotlib

```
x = np.arange(0, 10, 0.001)
y1 = np.exp(x)
y2 = -np.exp(x)
plt.plot(x, y1)
plt.plot(x, y2)
```

How to make **Legend**

```
x = np.arange(0, 10, 0.001)
y1 = np.exp(x)
y2 = -np.exp(x)
plt.plot(x, y1, label = "Positive")
plt.plot(x, y2, label = "Negative")
plt.legend(loc = "best")
```

How to add **axis description**

```
x = np.arange(0, 10, 0.001)
y1 = np.exp(x)
y2 = -np.exp(x)
plt.plot(x, y1, label = "Positive")
plt.plot(x, y2, label = "Negative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc = "best")
```

How to change **color and line style**

```
x = np.arange(0, 10, 0.001)
y1 = np.exp(x)
y2 = -np.exp(x)
plt.plot(x, y1, "--", color = "green", label = "Positive")
plt.plot(x, y2, "-.", color = "black", label = "Negative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc = "best")
```
